>Stokes drift velocity

$$ u_{st}={H_s^2\omega k \over 16} {cosh[2k(z+h)] \over sinh^2(kh)}\hat k $$

> Hs = significant Wave Height
> Omega = angular wave frquemcy
> u_st = (u_st, v_st)
> k = wave number

Where k = 2$$\pi$$ / $$\lambda$$
  



```python
import netCDF4
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import datetime as dt
import time
import sys
import numpy as np
import pandas as pd
```

```python
pickle = 0
if not pickle:
    dateInput = "2010-01-01 00:00" # Get Data for this day -> Currently plotting the whole year
    inputFloat = mdates.datestr2num(dateInput) #Make into datetime float
    inputDateTime = mdates.num2date(inputFloat)

    year = inputDateTime.year
    # This url points to the netCDF3 file on openDap server.
    # data_url = "http://dods.ndbc.noaa.gov/thredds/dodsC/data/stdmet/46092/46092h" + str(year)  +".nc"

    data_url = "http://dods.ndbc.noaa.gov/thredds/dodsC/data/stdmet/46042/46042h" + str(year)  +".nc"
    try:
        nc = netCDF4.Dataset(data_url)
    except RuntimeError:
        print("Couldn't find %s" % (data_url))
        print "Check internet connection or date"
        sys.exit()
    # Print out the variable names of the netCDF
    # for var in nc.variables:
    #     print var
    time = nc["time"]
    floatTime = []
    for date in mdates.epoch2num(time[:]):# Convert epoch to numPy float
        floatTime.append(date)
        # Tabularize the data into a pandas dataFrame
    # Make a datafram out of the netcdf file
    df = pd.DataFrame()
    for colName in nc.variables:
        if nc[colName].shape[0] > 1:
            if len(nc[colName].shape) == 1: 
                df[colName] = nc[colName][:][:]
            else:
                tempArray = []
                for dat in nc[colName][:]: # This loop
                    tempArray.append(dat[0][0])
                df[colName] = tempArray
    df['time'] = map(dt.datetime.fromtimestamp,nc['time']) # Convert time pt
    df.index = df['time'] # Set the index to time
    df.to_pickle('NBDC.pkl')
else:
    df = pd.read_pickle('NBDC.pkl')
```

```python
# Issue: The data is stored in a masked narray that has three dimensions
# but it appears that each x value only has one value, so reshaping may be in order
sstData = df['sea_surface_temperature']
# Reshape the array to be in one dimension
sstPltData  = sstData.reshape(-1)
# Remask the data array and the time array.
# Any value over 50 degrees is safely bad data and should be ignored.
sstMasked = np.ma.masked_where(sstPltData > 50, sstPltData)
# sstPltData > 50 is used to qualify if a data point is valid and whether it should be masked or not
# Using this argument for the floatTime array will ensure that date/time and the sst are properly aligned
timeMasked = np.ma.masked_where(sstPltData > 50, floatTime)
```

```python
# Add logic to check that speed and dir are the same size

xUDir = []
yUDir = []
for i in range(len(df['wind_dir'])):
    xSpeed = df['wind_spd'][i] * np.cos(np.deg2rad(df['wind_dir'][i]))
    ySpeed = df['wind_spd'][i] * np.sin(np.deg2rad(df['wind_dir'][i]))
    xUDir.append(xSpeed)
    yUDir.append(ySpeed)
```

```python
# Low pass filter the data with a buttersworth
# Based on http://stackoverflow.com/questions/25191620/creating-lowpass-filter-in-scipy-understanding-methods-and-units
from scipy.signal import butter, lfilter, freqz

def butter_lowpass(cutoff, fs, order=5):
    nyq = 0.5 * fs
    normal_cutoff = cutoff / nyq
    b, a = butter(order, normal_cutoff, btype='low', analog=False)
    return b, a

def butter_lowpass_filter(data, cutoff, fs, order=12):
    b, a = butter_lowpass(cutoff, fs, order=order)
    y = lfilter(b, a, data)
    return y

fs = 1. / (np.mean(np.diff(uTimeMasked[:10])) * 86400)
cutoff = 1. /(3600*20)
order = 1
y = butter_lowpass_filter(uSpeedMasked, cutoff, fs, order) #Return the data
b, a = butter_lowpass(cutoff, fs, order)
# Plot the frequency response.
w, h = freqz(b, a, worN=8000)
plt.subplot(1, 1, 1)
plt.plot(0.5*fs*w/np.pi, np.abs(h), 'b')
plt.plot(cutoff, 0.5*np.sqrt(2), 'ko')
plt.axvline(cutoff, color='k')
plt.xlim(0, 0.5*fs)
plt.title("Lowpass Filter Frequency Response")
plt.xlabel('Frequency [Hz]')
plt.grid()

def movingaverage(interval, window_size):
    window= np.ones(int(window_size))/float(window_size)
    return np.convolve(interval, window, 'same')
moving_Speed = movingaverage(uSpeedMasked, 12)
```

```python
from matplotlib.dates import date2num
def stick_plot(time, u, v, **kw):
    width = kw.pop('width', 0.002)
    headwidth = kw.pop('headwidth', 0)
    headlength = kw.pop('headlength', 0)
    headaxislength = kw.pop('headaxislength', 0)
    angles = kw.pop('angles', 'uv')
    ax = kw.pop('ax', None)
    
    if angles != 'uv':
        raise AssertionError("Stickplot angles must be 'uv' so that"
                             "if *U*==*V* the angle of the arrow on"
                             "the plot is 45 degrees CCW from the *x*-axis.")

    time, u, v = map(np.asanyarray, (time, u, v))
    if not ax:
        fig, ax = plt.subplots()
    
    q = ax.quiver(time, [[0]*len(time)], u, v,
                  angles='uv', width=width, headwidth=headwidth,
                  headlength=headlength, headaxislength=headaxislength,
                  **kw)

    ax.axes.get_yaxis().set_visible(False)
    ax.xaxis_date()
    return q

fig, ax = plt.subplots(figsize=(11, 2.75))

q = stick_plot(uTimeMasked, xUDir, yUDir, ax=ax, width=0.002, color='green')

ref = 1
qk = ax.quiverkey(q, 0.1, 0.85, ref,
                  "%s M s$^{-1}$" % ref,
                  labelpos='N', coordinates='axes')
_ = plt.xticks(rotation=70)
```
